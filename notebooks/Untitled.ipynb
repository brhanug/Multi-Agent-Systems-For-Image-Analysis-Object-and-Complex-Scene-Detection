```python
import pandas as pd

base_csv = "/home/brhanu/thesis_project/runs/detect/train_baseline/results.csv"
pseudo_csv = "/home/brhanu/thesis_project/runs/detect/train_refine_long3/results.csv"

df_base = pd.read_csv(base_csv)
df_pseudo = pd.read_csv(pseudo_csv)

# 🔹 Clean up column names
df_base.columns = df_base.columns.str.strip()
df_pseudo.columns = df_pseudo.columns.str.strip()

# 🔹 Select relevant metrics
metrics = ['metrics/precision(B)', 'metrics/recall(B)', 'metrics/mAP50(B)']

print("=== Baseline YOLO ===")
print(df_base[metrics].iloc[-1])
print("\n=== Pseudo-Labeled YOLO ===")
print(df_pseudo[metrics].iloc[-1])
```

```python
import matplotlib.pyplot as plt, cv2, os

base_dir = "/home/brhanu/thesis_project/runs/compare/baseline_preds"
pseudo_dir = "/home/brhanu/thesis_project/runs/compare/pseudo_preds"
sample = "00000163_0.jpg"  # replace with any validation image

img_base = cv2.cvtColor(cv2.imread(os.path.join(base_dir, sample)), cv2.COLOR_BGR2RGB)
img_pseudo = cv2.cvtColor(cv2.imread(os.path.join(pseudo_dir, sample)), cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1); plt.imshow(img_base); plt.title("Baseline"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(img_pseudo); plt.title("Pseudo-Labeled"); plt.axis("off")
plt.show()
```
